```python
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
train_df = pd.read_csv('spam_train.csv')
test_df = pd.read_csv('spam_test.csv')
```

```python
test_df = test_df.drop(' ID', axis = 1)
```

```python
X_train, y_train, X_test, y_test = train_df.iloc[:, :-1], train_df.iloc[:, -1], test_df.iloc[:, :-1], test_df.iloc[:, -1]
```

```python
def euc_distance(train, test):
    
    final = []
    
    for i in range(len(train)):
        summation = []
        distance = []
        for j in range(len(train.columns)):
            summation.append((test[j] - train.iloc[i, j]) ** 2)

            # distances.append(summation)
        distance.append([i, np.sum(summation) ** 0.5])
        final.append(distance)
        # final[i] = distance

    # return [final]
    return [final]
```

```python
train_df['distance'] = 0

for i in tqdm(range(len(train_df)), desc="Calculating distances"):
    train_df.loc[i, 'distance'] = euc_distance(X_train, X_test.iloc[i, :])

train_df
```

```python
train_df.to_csv('distances_calculated.csv')
```
